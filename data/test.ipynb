```python
import pandas as pd
```

```python
df = pd.read_csv('raw/top_10000_1960-now.csv')
```

```python
df.head()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.columns = df.columns.str.replace('(', '').str.replace(')', '')
df = df.drop(columns=['added_at', 'added_by'])
df.head()
```

```python
df = df.rename(columns={'track_uri': 'track_id','artist_uris': 'artist_id','album_uri':'album_id', 'album_artist_uris':'album_artist_id'})

df['album_id'] = df['album_id'].str.replace('spotify:album:', '', regex=False)
df['album_artist_id'] = df['album_artist_id'].str.replace('spotify:artist:', '', regex=False)
df['track_id'] = df['track_id'].str.replace('spotify:track:', '', regex=False)
df.head()
```

```python
#df = df.set_index('track_id')
df = df.drop(columns=['track_preview_url', 'album_genres', 'copyrights'])
```

```python
df.head()
```

```python
for i in df['artist_id']:
    if i != '':
        df['artist_id'] = df['artist_id'].str.replace('spotify:artist:', '', regex=False)
        df['artist_id'] = df['artist_id'].str.replace(' ', '', regex=False)

df.head()
```

```python
df.head(10)
```

```python
df1 = df.dropna(subset=['track_id', 'track_name', 'artist_id', 'artist_names', 'album_release_date', 'album_id','album_name', 'album_artist_id','album_artist_names', 'album_image_url'])
```

```python
df1.head(10)
```

```python
def standardise_date(date_str):
    if pd.isna(date_str) or date_str == '':
        return pd.NaT

    formats = [
        '%Y/%m/%d', '%Y-%m-%d', '%d %b %Y', '%b %d, %Y', '%d %B %Y',
        '%d-%m-%Y', '%d/%m/%Y'
    ]
    for fmt in formats:
        try:
            return pd.to_datetime(date_str, format=fmt)
        except ValueError:
            continue

    return pd.NaT
```

```python
df1['album_release_date'] = (df1['album_release_date'].apply(standardise_date))
df1['album_release_date'] = (df1['album_release_date'].dt.strftime('%d/%m/%Y'))
```

```python
df1.head()
```

```python
df1 = df1.dropna(subset=['album_release_date'])
df1.head()
```

```python
df1['album_release_date'] = pd.to_datetime(df1['album_release_date'], errors='coerce')
df1['album_release_date'] = df1['album_release_date'].dt.strftime('%d/%m/%Y')
```
